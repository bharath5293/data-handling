```python
import requests
Link="https://en.wikipedia.org/wiki/List_of_Indians_by_net_worth"
Link_text =requests.get(Link).text
print(Link_text)
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(Link_text,'lxml')
print(soup)
```

```python


import requests
from bs4 import BeautifulSoup

# URL of the Wikipedia page for India's richest women
url = 'https://en.wikipedia.org/wiki/List_of_female_billionaires'

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the table containing the list of billionaires
    table = soup.find('table', class_='wikitable sortable')

    # Initialize an empty list to store the richest women
    richest_women = []

    # Iterate over each row in the table (skipping the header row)
    for row in table.find_all('tr')[1:]:
        # Extract the columns of each row
        columns = row.find_all('td')

        # Extract the name of the woman
        name = columns[1].text.strip()

        # Extract the net worth of the woman
        net_worth = columns[3].text.strip()

        # Append the name and net worth to the list
        richest_women.append((name, net_worth))

    # Print the list of richest women
    print("India's Richest Women:")
    for idx, (name, net_worth) in enumerate(richest_women, start=1):
        print(f"{idx}. {name} - {net_worth}")
else:
    print("Failed to retrieve the webpage. Please check the URL or your internet connection.")
```
