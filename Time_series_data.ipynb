```python
import pandas as pd
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
```

```python
data={
    "date":pd.to_datetime(["2022-1-5","2022-02-12","2022-3-1","2022-4-10","2022-5-1","2022-6-1","2022-7-16","2022-8-5","2022-9-1","2022-10-3",]),
    "value":[10,15,18,19,21,25,29,35,39,50]
}
df=pd.DataFrame(data)
df.set_index("date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["value"],marker='o')
plt.xlabel("date")
plt.ylabel("value")
plt.title("time series with trend")
plt.grid(True)
plt.show()
```

```python
data={
    "date":pd.to_datetime(["2022-1-5","2022-02-12","2022-3-1","2022-4-10","2022-5-1","2022-6-1","2022-7-16","2022-8-5","2022-9-1","2022-10-3",]),
    "value":[100,150,128,200,180,160,170,300,202,310]
}
df=pd.DataFrame(data)
df.set_index("date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["value"])
plt.xlabel("date")
plt.ylabel("value")
plt.title("time series with seasonality")
plt.grid(True)
plt.show()
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/sunspotarea.csv', parse_dates=['date'], index_col='date')
result = seasonal_decompose(data, model='additive', period=12)


fig, axes = plt.subplots(1, 3, figsize=(20, 4), dpi=100)


data.plot(ax=axes[0], title='Original Data', legend=False)


result.trend.plot(ax=axes[1], title='Trend Only', legend=False)


result.seasonal.plot(ax=axes[2], title='Seasonal Only', legend=False)

plt.tight_layout()
plt.show()
```
